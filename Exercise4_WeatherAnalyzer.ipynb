```python
import requests
import csv

# Fetch weather data function
def fetch_weather(city, api_key):
    try:
        url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print("Error fetching weather:", e)
        return None

# Analyze weather data
def analyze_weather(weather_data):
    temp = weather_data["main"]["temp"]
    wind = weather_data["wind"]["speed"]
    humidity = weather_data["main"]["humidity"]

    # Temperature category
    if temp <= 10:
        summary = "Cold (≤10°C)"
    elif 11 <= temp <= 24:
        summary = "Mild (11-24°C)"
    else:
        summary = "Hot (≥25°C)"

    # Add warnings
    if wind > 10:
        summary += " | High wind alert!"
    if humidity > 80:
        summary += " | Humid conditions!"

    return summary

# Log weather data to CSV
def log_weather(city, filename):
    api_key = "YOUR_API_KEY_HERE"
    data = fetch_weather(city, api_key)

    if data:
        summary = analyze_weather(data)

        with open(filename, "a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([
                city,
                data["main"]["temp"],
                data["wind"]["speed"],
                data["main"]["humidity"],
                summary
            ])

        print("Weather data logged successfully.")

# Example run
log_weather("London", "weather_log.csv")
```
